```python
import pandas as pd
import torch 
from torch.utils.data import DataLoader, TensorDataset, random_split
import matplotlib.pyplot as plt
```

```python
device = torch.device("cpu")
```

```python
device
```

```python
data = pd.read_csv("Data/train.csv")
```

```python
if "id" in data.columns and "day" in data.columns:
    
    data.drop(["id", "day"], axis = 1, inplace = True)
```

```python
data_tensor = torch.tensor(data.values, dtype = torch.float32, device = device)
```

```python
X = data_tensor[:, :-1]
y = data_tensor[:, -1].unsqueeze(1)
```

```python
final_data = TensorDataset(X,y)

train_size = int(0.80 * len(final_data))
test_size = len(final_data) - train_size

train_data, test_data = random_split(final_data, [train_size, test_size])

print(f"Train Size = {train_size}\nTest Size = {test_size}")
```

# Standardize

```python
train_x = torch.stack([sample[0] for sample in train_data])

mean = train_x.mean()
std = train_x.std()

def standarization(tensor):
    return (tensor - mean)/std

train_data = [(standarization(X), y) for X, y in train_data]

    
```

```python
train_x
```

```python
mean = sum(x for x, _ in train_data)/len(train_data)
std = torch.sqrt( sum((x - mean)**2 for x, _ in train_data) / len(train_data) )


train_data = [((x - mean)/std, y ) for x, y in train_data ]
test_data = [((x- mean)/std, y) for x, y in test_data ] 
```
